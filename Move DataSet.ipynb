```python
#Checking column wise null values percentage

null_percent = (movie_df.isnull().sum()/len(movie_df))*100
null_percent
```

```python
#As we will be deailing mostly with columns like ratings, gross collection, popularity of movies , let us drop other unwanted columns

new_movie_df = movie_df.drop(['color','director_facebook_likes','actor_1_facebook_likes','actor_2_facebook_likes','actor_3_facebook_likes','actor_2_name','cast_total_facebook_likes','actor_3_name','duration','facenumber_in_poster','content_rating','country','movie_imdb_link','aspect_ratio','plot_keywords'],axis=1)
new_movie_df.head()
```

```python
#Now that we have relevant columns, let's delete rows from columns having high null percentage(>5%)
new_movie_df = new_movie_df[~np.isnan(new_movie_df['gross'])]
new_movie_df = new_movie_df[~np.isnan(new_movie_df['budget'])]
new_movie_df.head()
```

```python
#Dropping rows which have more than 5 null values
new_movie_df = new_movie_df[new_movie_df.isnull().sum(axis=1)<=5]
new_movie_df.head()
```

```python
#Filling missing languages column with english
new_movie_df['language'].fillna('English',inplace=True)
```

```python
#Checking number of rows left
new_movie_df.shape
```

```python
#Changing amount from dollars to million dollars
new_movie_df[['gross','budget']].apply(lambda x:x/1000000)
```

```python
new_movie_df['Profit'] = new_movie_df['gross'] - new_movie_df['budget']
```

```python
#Finding highest profit movies
new_movie_df.sort_values(by='Profit',ascending=False,inplace=True,)
```

```python
top10Profit = new_movie_df.head(10)
top10Profit[['movie_title','Profit']]
top10Profit.set_index('movie_title')
```

```python
#Dropping Duplicate Values
```

```python
new_movie_df.drop_duplicates()
```

```python
#Repeating the above task for finding top 10 profitable movies
newTop10 = new_movie_df.sort_values(by= 'Profit', ascending = False).head(10)
newTop10
```

```python
#Finding IMDB 250 Movies
#Filtering movies with more than 25000 votes and keeping in a dataset
imdbNominee = new_movie_df[new_movie_df.num_voted_users>25000]
imdbNominee.head(10)
```

```python
imdb250 = imdbNominee.sort_values(by ='imdb_score' , ascending = False).head(250)
imdb250.head(10)
imdb250['Rank'] = range(1,251)
imdb250.set_index('Rank')
```

```python
TopForeignLangFilms = imdb250[~(imdb250['language'] == 'English')]
TopForeignLangFilms.head()
```

```python
#Finding top 10 directors based on mean imdb score
Top10Director = new_movie_df.groupby('director_name')['imdb_score'].mean().sort_values(ascending=False).head(10)
Top10Director
```
